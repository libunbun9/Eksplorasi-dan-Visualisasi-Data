```python
# import "pandas" package and set the name as "pd"
import pandas as pd
```

```python
# import fire case in Jakarta from 2018
# Source : Satu Data Indonesia
data_kebakaran = pd.read_csv(r"C:\Users\yippi\Downloads\kebakaran_kuis2.csv")
```

```python
# call the dataset
data_kebakaran
```

```python
# subset data using iloc
data_loc1 = pd.DataFrame(data_kebakaran.iloc[range(0,10),[0,3]])
data_loc1
```

```python
# subset data using iloc 
data_loc2 = data_kebakaran.iloc[range(0,10),[0,10]]
data_loc2
```

```python
# # merge data first and second iloc from previous code, merge by "_id" index.
```

```python
merge_datailoc = pd.merge(data_loc1,data_loc2, on ="_id")
```

```python
# call merged dataset
merge_datailoc
```
