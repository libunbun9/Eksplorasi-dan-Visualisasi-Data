```python
# the first rime we do is import pandas package as pd
import pandas as pd
```

Data yang akan digunakan adalah data pengeluaran rata-rata perkapita per-bulan menurut golongan pengeluaran dan kelompok barang di Jakarta. Data bersumber dari https://www.data.go.id/ data provinsi Jakarta. Data akan diambil adalah data kelompok barang makanan.

```python
data_kapita = pd.read_csv(r"C:\Users\yippi\Downloads\Filedata Data Pengeluaran Rata-rata Perkapita Per Bulan Menurut Golongan Pengeluaran dan Kelompok Barang DKI Jakarta.csv")
```

```python
data_kapita # preview the data
```

```python
#because of golongan_pengeluaran <200.000 is 0, we can delete the row.
data_kapita.drop(data_kapita.index[0:2], inplace=True)
```

```python
data_kapita # show data after delete row 
```

```python
# Check data types of all columns
data_kapita.dtypes
```

Kategori jumlah masih dalam berbentuk data kategorik, maka akan diubah menjadi numeric (integer)

```python
# change jumlah ("sum") into numeric (int)
data_kapita['jumlah'] = data_kapita['jumlah'].astype(int)
```

```python
# jumlah already changed into numeric
data_kapita.dtypes
```

```python
subset_kapita = data_kapita[(data_kapita['kelompok_barang']=='Makanan')] # subset kelompok barang menjadi makanan saja
```

```python
subset_kapita # preview subseted data
```

Setelah data sudah rapi, maka data siap dibuatkan chart. Chart akan dibuat menggunakan package mathplotlib.

```python
# import matplotlib.pyplot to make a plot, we can call that library as plt.
import matplotlib.pyplot as plt
```

```python
# bar plot

gp = subset_kapita['golongan_pengeluaran']
jumlah = subset_kapita['jumlah']

plt.bar(gp, jumlah)
plt.xlabel('Golongan Pengeluaran (Rp)')
plt.ylabel('Banyaknya')
plt.title('Banyaknya Orang berdasarkan Pengeluaran untuk Makanan pada Tahun 2014')
plt.show()
```

```python
# line chart
plt.plot([200000,300000,500000,700000,1000000], subset_kapita['jumlah'], marker='o', linestyle='--', color='blue')
plt.title('Banyaknya Orang Berdasarkan Pengeluaran untuk Makanan Tahun 2014')
plt.ylabel('Banyaknya')
plt.xlabel('Rupiah (Rp)')
plt.show()
```

```python
# pie chart
plt.figure(figsize=(8, 8))  # Adjusting the size of our pie (optional)
plt.pie(jumlah, labels=gp, autopct='%1.1f%%', startangle=90)
plt.title('Pengeluaran Rata-rata untuk Makan Tahun 2014')  # Adding a title to our pie chart
plt.show()
```

```python
# histogram

plt.hist(jumlah)
plt.xlabel('Jumlah Pengeluaran Perkapita (Rp)')
plt.title('Histogram of Exam Scores')
plt.show()
```

```python
# make scatter plot data
subset_kapita2 = data_kapita[(data_kapita['kelompok_barang']=='Bukan Makanan')] # subset kelompok barang menjadi bukan makanan
```
