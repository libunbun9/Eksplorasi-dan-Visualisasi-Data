## Checking The Distribution
Untuk mengecek distribusi, yang akan digunakan adalah metode visual dan tes. Metode visual yang digunakan adalah histogram, boxplot dan QQ plot. Metode test yang digunakan adalah Shapiro-Wilk Test.

```python
# Using histogram
plt.hist(belanja_negara)
plt.title('Distribusi Anggaran Belanja Negara')
plt.xlabel('Rp')
plt.ylabel('Frekuensi')
plt.show()
```

berdasarkan histogram di atas, dapat terlihat bahwa data anggaran tidak berdistribusi normal. Data dikatakan tidak berdistribusi normal ketika grafik hitogram tidak membentuk lonceng bel. Dari grafik ini bisa dikatakan bahwa data negative skewed atau data lebih condong ke kiri. Sepertinya nanti akan ada pencilan karena diantara Rp7.500.000.000.000 - 1.500.00.00.00.000 tidak memiliki frekuensi satupun pada data anggaran kali ini.

```python
# using boxplot
plt.boxplot(belanja_negara)
plt.title("Anggaran Belanja Negara")
plt.show()
```

Dari hasil box plot rupanya terdapat 5 kota yang termasuk pencilan. Jika dilihat dari box dan whiskernya, nilai box mendekati min dari data, maka data negative skewed. Dilihat dari boxplot tersebut bisa disimpulkan bahwa data tidak berdistribusi normal.

```python
# using QQ plot
stats.probplot(belanja_negara, dist="norm", plot=plt)
plt.title('QQ-plot Anggaran Belanja Negara')
plt.show()
```

Selanjutnya, untuk memastikan bahwa data tidak berdistribusi normal, digunakanlah qqplot untuk melihat kenormalan data. Data dikatakan normal ketika titik" plot mengikuti garis. Pada qqplot yang dihasilkan, titik dan garis tidak mengikuti satu sama lain. Sehingga data tidak berdistribusi normal.

Untuk menguji atau emmastikan lagi apakah benar data tidak berdistribusi normal, digunakan uji shapiro-wilk. Pada uji shapiro-wilk hipotesis yang digunakan adalah sebagai berikut:<br>
- H0 : Data berdistribusi normal
- H1 : Data tidak berdistribusi normal
<br>
Ketika p-value>H0, maka H0 diterima, data berdistribusi normal.

```python
# Using Shapiro-Wilk test
from scipy.stats import shapiro

stat, p_value = shapiro(belanja_negara)

print(f'Statistik Uji: {stat}, p-value: {p_value}')
if p_value > 0.05:
    print("Data kemungkinan berdistribusi normal")
else:
    print("Data tidak berdistribusi normal")
```

didapatkan bawa pvalue < alpha 0.05, sehingga data tidak berdistribusi normal.
